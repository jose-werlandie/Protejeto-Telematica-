```python
import requests
import pandas as pd
from IPython.display import display, Markdown
```

```python
url = "http://api.portaldatransparencia.gov.br/api-de-dados/despesas/documentos-por-favorecido"
```

ano *
integer($int32)
(query)	
Ano de emissão do documento

chave-api-dados *
string
(header)	
Chave para acessar à API. Para obter a chave acesse http://www.portaldatransparencia.gov.br/api-de-dados/cadastrar-email

codigoPessoa *
string
(query)	
Código do Favorecido (CPF, CNPJ ou código do SIAFI)

fase *
integer($int32)
(query)	
Fase da despesa (1 - Empenho, 2 - Liquidação, 3 - Pagamento)

gestao
string
(query)	
Código da gestão do documento

pagina *
integer($int32)
(query)	
Página consultada

Default value : 1

```python
DataMes = input("mes : ")
ano = 2019
headers = {"chave-api-dados" : "xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"}
codigoPessoa (CPF) = "xxxxxxxxxxx"
fase = 3
pagina = 1
```

```python
params = {'ano':ano, 'codigoPessoa':codigoPessoa, 'fase':fase , 'pagina':pagina}
```

```python
resultado = requests.get(url, params=params, headers=headers)
resultado
```

```python
nome = resultado.json()
```

```python
def render(ndocumento, data, descricao, fase, tipodocumento, valordocumento, obs):
    
    Ndocumento ='<h6><section class="dados-tabelados"><div class="row"><div class="col-xs-12 col-sm-3"><strong>Nº do documento</strong><span><p>'+ndocumento+'</p></span></div>'
    Data = '<div class="col-xs-12 col-sm-3"><strong>Data</strong><span><p>'+data+'</p></span></div>'
    Descricao = '<div class="col-xs-12 col-sm-6"><strong>Descrição</strong><span><p>'+descricao+'</p></span></div></div><br>'
    Fase = '<div class="row"><div class="col-xs-12 col-sm-3"><strong>Fase</strong><span><p>'+fase+'</p></span></div>'
    TipoDocumento = '<div class="col-xs-12 col-sm-3"><strong>Tipo de documento</strong><span><p>'+tipodocumento+'</p></span></div>'
    ValorDocumento = '<div class="col-xs-12 col-sm-6"><strong>Valor do documento</strong><span><p>'+valordocumento+'</p></span></div></div><br>'
    Observacao = '<div class="row"><div class="col-xs-12"><strong>Observação do documento</strong><span><p>'+obs+'</p></span></div></div></section><h6>'
    
    display(Markdown(Ndocumento+Data+Descricao+Fase+TipoDocumento+ValorDocumento+Observacao))
```

```python
for i in nome:
    mes = i['data'].split("/")[1]
    if mes == DataMes:
        ndocumento=i['documentoResumido']
        data=i['data']
        descricao='Ordem Bancária (OB)'
        fase=i['fase']
        tipodocumento='OBC/OBB PARA TERCEIROS EM OUTROS BANCOS'
        valordocumento=i['valor']
        obs=i['observacao']
        render(ndocumento, data, descricao, fase, tipodocumento, valordocumento, obs)
        
```
